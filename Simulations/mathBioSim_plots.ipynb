```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("HM_2D.dat.csv", sep=';', header=None, skiprows=1)
data.columns = ['sd_w_in', 'sd_w_ex', 'N 1', 'N 2', 'time1', 'time2']

plt.title('Proportions of N1/N2 for HM 2D')
plt.tricontourf(data['sd_w_in'], data['sd_w_ex'], data['N 1']/data['N 2'], camp='Greens', label='N1/N2', levels=25)
plt.colorbar()
plt.xlabel('sd_w_in')
plt.ylabel('sd_w_ex')
plt.legend()
plt.show()

plt.title("coexistence of N1, HM 2D")
plt.scatter(data.loc[data['N 1'] == 0, 'sd_w_ex'], data.loc[data['N 1'] == 0, 'sd_w_in'], color='red', label='N 1 = 0')
plt.scatter(data.loc[data['N 1'] != 0, 'sd_w_ex'], data.loc[data['N 1'] != 0, 'sd_w_in'], color='blue', label='N 1 != 0')
plt.ylabel('sd_w_in')
plt.xlabel('sd_w_ex')
plt.legend(["Not coexistence", "Coexistence"], loc=2)
plt.show()


plt.title("coexistence of N2, HM 2D")
plt.scatter(data.loc[data['N 2'] == 0, 'sd_w_ex'], data.loc[data['N 2'] == 0, 'sd_w_in'], color='red', label='N 2 = 0')
plt.scatter(data.loc[data['N 2'] != 0, 'sd_w_ex'], data.loc[data['N 2'] != 0, 'sd_w_in'], color='blue', label='N 2 != 0')
plt.ylabel('sd_w_in')
plt.xlabel('sd_w_ex')
plt.legend(["Not coexistence", "Coexistence"], loc=2)
plt.show()

plt.title('2D HM, number of type 1')
plt.tricontourf(data['sd_w_in'], data['sd_w_ex'], data['N 1'], cmap='Blues', label='N 1', levels = 25)
plt.colorbar()
plt.xlabel('sd_w_in')
plt.ylabel('sd_w_ex')
plt.legend()
plt.show()

plt.title('2D HM, number of type 2')
plt.tricontourf(data['sd_w_in'], data['sd_w_ex'], data['N 2'], cmap='Reds', label='N 2', levels = 25)
plt.colorbar()
plt.xlabel('sd_w_in')
plt.ylabel('sd_w_ex')
plt.legend()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('CCTO_1D.csv')
data['d12'] = data['d12'].replace(',', '.', regex=True).astype(float)

plt.title("graph of coexistence")
plt.scatter(data.loc[data['N 1'] == 0, 'd12'], data.loc[data['N 1'] == 0, 'sd_m_2'], color='red', label='N 1 = 0')
plt.scatter(data.loc[(data['N 1'] != 0) & (data['N 2'] != 0), 'd12'], data.loc[(data['N 1'] != 0) & (data['N 2'] != 0), 'sd_m_2'], color='yellow', label='N1 != 0 && N2 !=0')
plt.scatter(data.loc[data['N 2'] == 0, 'd12'], data.loc[data['N 2'] == 0, 'sd_m_2'], color='black', label='N 2 = 0')
plt.ylabel('sd_m_2')
plt.xlabel('d12')
plt.legend(["Victim","Coexistence", "Predator"], loc=1)
plt.show()

plt.title("graph of coexistence with population proportion N1/N2")
plt.scatter(data['d12'], data['sd_m_2'], c=data['N 1']/data['N 2'], cmap='Blues', label='N1/N2')
plt.colorbar()
plt.xlabel('sd_w_in')
plt.ylabel('sd_w_ex')
plt.legend(loc=2)
plt.show()

plt.title("graph of coexistence with population proportion N2/N1")
plt.scatter(data['d12'], data['sd_m_2'], c=data['N 2']/data['N 1'], cmap='Blues', label='N2/N1')
plt.colorbar()
plt.xlabel('sd_w_in')
plt.ylabel('sd_w_ex')
plt.legend(loc=2)
plt.show()

plt.title("one-dimensional CCTO case (Number of the first type)")
plt.tricontourf(data['sd_m_2'], data['d12'], data['N 1'], cmap='Greens', label='N 1')
plt.colorbar()
plt.ylabel('d12')
plt.xlabel('sd_m_2')
plt.legend(["Competition of N1"], loc=2)
plt.show()

plt.title("one-dimensional CCTO case (Number of the second type)")
plt.tricontourf(data['sd_m_2'], data['d12'], data['N 2'], cmap='Greens', label='N2')
plt.colorbar()
plt.ylabel('d12')
plt.xlabel('sd_m_2')
plt.legend(["Competition of N2"], loc=2)
plt.show()

data['time1'] = data['time1'].replace(',', '.', regex=True).astype(float)

plt.scatter(data['d12'], data['sd_m_2'], c=data['time1'], cmap='Purples', label='time1')
plt.colorbar()
plt.xlabel('d12')
plt.ylabel('sd_m_2')
plt.legend(loc=2)
plt.show()


data['time2'] = data['time2'].replace(',', '.', regex=True).astype(float)

plt.scatter(data['d12'], data['sd_m_2'], c=data['time2'], cmap='Purples', label='time2')
plt.colorbar()
plt.xlabel('d12')
plt.ylabel('sd_m_2')
plt.legend(loc=2)
plt.show()
```
